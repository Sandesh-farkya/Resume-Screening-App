```python
import re
def cleanResume(txt):
    cleaned_txt = re.sub(r'http\S+\s+', ' ', txt)  # Remove extra spaces
    cleaned_txt = re.sub(r'www\S+\s+', ' ', cleaned_txt)  # Remove extra spaces
    cleaned_txt = re.sub(r'\S+@\S+', ' ', cleaned_txt)  # Remove email addresses
    cleaned_txt = re.sub('[%s]' % re.escape(r'!"#$%&\'()*+,-./:;<=>?@[\\]^_`{|}~'), ' ', cleaned_txt)  # Remove punctuation
    cleaned_txt = re.sub(r'[^a-zA-Z\s]', ' ', cleaned_txt)  # Remove special characters
    cleaned_txt = re.sub(r'\s+', ' ', cleaned_txt)  # Remove extra spaces
    cleaned_txt = cleaned_txt.strip()  # Remove leading and trailing spaces
    return cleaned_txt
```

```python
data["Resume"] = data["Resume"].apply(lambda x: cleanResume(x))
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
data["Category"] = le.fit_transform(data["Category"])
data["Category"].unique()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf.fit_transform(data["Resume"])
tfidf_matrix.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(tfidf_matrix, data["Category"], test_size=0.2, random_state=42)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.multiclass import OneVsRestClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
knn = OneVsRestClassifier (KNeighborsClassifier(n_neighbors=5))
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)
print("Accuracy:", accuracy_score(y_test, y_pred))
```

```python
import pickle
pickle.dump(knn, open('knn_model.pkl', 'wb'))
pickle.dump(tfidf, open('tfidf_vectorizer.pkl', 'wb'))
loaded_model = pickle.load(open('knn_model.pkl', 'rb'))
```
