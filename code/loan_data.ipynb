```python
import pandas as pd
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
loan_data=pd.read_csv("C:\\Users\\sporw\\Desktop\\python\\loan_prediction (1).csv")

# loan_data.drop(['Loan_ID'],axis=1,inplace=True)
# print(loan_data.info()) 
# loan_data.isnull().sum()
# loan_data[['Dependents','Self_Employed']].mode()
# loan_data[['LoanAmount','Loan_Amount_Term','Credit_History']].mean()
loan_data.fillna({
    'Gender': 'Male',
    'Married': 'Yes',
    'Dependents': 0,
    'Self_Employed': 'No',
    'LoanAmount': 146,
    'Loan_Amount_Term': 342,
    'Credit_History': 0
}, inplace=True)
loan_data.isnull().sum()
# print(loan_data)
# print(loan_data.info())
# sns.countplot(loan_data['Loan_Status'])


# loan_data.to_csv('edited_file.csv', index=False) 
```

```python
# Create subplots for multiple visualizations
fig, axes = plt.subplots(3, 2, figsize=(15, 14))

# Plot 1: Loan Status Distribution
sns.countplot(data=loan_data, x='Loan_Status', ax=axes[0, 0], palette='viridis')
axes[0, 0].set_title('Loan Status Distribution')
axes[0, 0].set_xlabel('Loan Status')
axes[0, 0].set_ylabel('Count')

# Plot 2: Loan Status by Gender
sns.countplot(data=loan_data, x3='Gender',hue='Loan_Status', ax=axes[0, 1], palette='viridis')
axes[0, 1].set_title('Loan Status by Gender')
axes[0, 1].set_xlabel('Gender')
axes[0, 1].set_ylabel('Count')

# Plot 3: Applicant Income Distribution
sns.histplot(data=loan_data, x='ApplicantIncome', kde=True, ax=axes[1, 0], color='blue', bins=30)
axes[1, 0].set_title('Applicant Income Distribution')
axes[1, 0].set_xlabel('Applicant Income')
axes[1, 0].set_ylabel('Frequency')

# Plot 4: Loan Amount vs Loan Status
sns.boxplot(data=loan_data, x='Loan_Status', y='LoanAmount', ax=axes[1, 1], palette='viridis')
axes[1, 1].set_title('Loan Amount by Loan Status')
axes[1, 1].set_xlabel('Loan Status')
axes[1, 1].set_ylabel('Loan Amount')

# Plot 5: Loan Status by Property Area
sns.countplot(data=loan_data, x='Property_Area', hue='Loan_Status', ax=axes[2, 0], palette='viridis')
axes[2, 0].set_title('Loan Status by Property Area')
axes[2, 0].set_xlabel('Property Area')
axes[2, 0].set_ylabel('Count')

# Plot 6: Credit History Distribution by Loan Status
sns.countplot(data=loan_data, x='Credit_History', hue='Loan_Status', ax=axes[2, 1], palette='viridis')
axes[2, 1].set_title('Credit History by Loan Status')
axes[2, 1].set_xlabel('Credit History')
axes[2, 1].set_ylabel('Count')

plt.tight_layout()
plt.show()
```
