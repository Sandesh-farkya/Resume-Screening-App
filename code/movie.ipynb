```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create a 2x2 subplot
fig, axs = plt.subplots(2, 2, figsize=(12, 8))

# Pie Chart
axs[0, 0].pie(updat_data['Country'].value_counts(), 
              labels=updat_data['Country'].value_counts().index, 
              autopct='%1.1f%%', 
              startangle=90)
axs[0, 0].set_title('Country Distribution')

# Bar Chart (Countplot)
sns.countplot(x='Primary Platform', 
              data=updat_data, 
              palette=["black", "red", "green", "blue", "skyblue"], 
              ax=axs[0, 1])
axs[0, 1].set_title('Primary Platform Distribution')
axs[0, 1].tick_params(axis='x', rotation=45)  # Rotate x-axis labels

# Add more visualizations to axs[1, 0] and axs[1, 1] if needed

axs[1, 0].set_title('Count of Primary Platform Usage by Country', fontsize=16)
sns.barplot(
    x='Country', 
    y='Count', 
    hue='Primary Platform', 
    data=updat_data.groupby(['Country', 'Primary Platform']).size().reset_index(name='Count'),
    palette='Set2'
)
plt.title('Count of Primary Platform Usage by Country', fontsize=16)
plt.xlabel('Country', fontsize=12)
plt.ylabel('Count', fontsize=12)

# Rotate x-axis labels for better readability
plt.xticks(rotation=45,color='red')

# Adjust layout and show the plot
plt.tight_layout()
plt.show()
axs[1, 1].axis('off')  # Placeholder for an empty subplot

# Adjust layout
plt.tight_layout()
plt.show()
```
